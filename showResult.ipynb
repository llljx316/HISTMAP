```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from pathlib import Path
import pickle
import textwrap

# 图片路径列表
root_path = Path('result/segres/Bodleian Library/ec925704-e84d-4180-8214-a9737fb5d993')
image_paths = list(root_path.glob('*.png'))
text_res_path = root_path / 'text_res.pickle'
with open(text_res_path, 'rb') as tpickle:
    text_dict = pickle.load(tpickle)
# image_paths = ['path/to/image1.png', 'path/to/image2.png', 'path/to/image3.png']
# letters = ['A', 'B', 'C']  # 每张图片对应的字母

# 显示多张图片并添加字母说明
fig, axes = plt.subplots(1, int(len(image_paths)), figsize=(15, 5))
for i, ax in enumerate(axes):
    try:
        ax.axis('off')
        img = mpimg.imread(image_paths[i])
        ax.imshow(img)
        # 获取图片的名称并查找对应的文字说明
        image_name = Path(image_paths[i]).stem
        caption = text_dict.get(image_name, "")

        # 使用 textwrap.wrap 将文字分行
        wrapped_caption = "\n".join(textwrap.wrap(image_paths[i].stem+caption, 15))

        # 设置标题，限制字体大小
        ax.set_title(wrapped_caption, fontsize=8)  # 设置小字体大小
    except:
        continue
    

plt.show()
```
